The notebook will demonstrate four join types: INNER JOIN, LEFT JOIN, RIGHT JOIN, and FULL OUTER JOIN. The results produced by each join will be examined and compared in order to understand how they differ and what type of records each join returns.

## Preview of Questions Table

The query below displays a sample of the `posts_questions` table.

```python
%%bigquery --project quiz11-joins

SELECT *
FROM `bigquery-public-data.stackoverflow.posts_questions`
LIMIT 10;
```

## Preview of the Answers Table

The query below displays a sample of the `posts_answers` table.

```python
%%bigquery --project quiz11-joins

SELECT *
FROM `bigquery-public-data.stackoverflow.posts_answers`
LIMIT 10;
```

## INNER JOIN

This query returns only questions that have matching answers.

```python
%%bigquery --project quiz11-joins

SELECT
q.id AS question_id,
q.title,
a.id AS answer_id,
a.score
FROM `bigquery-public-data.stackoverflow.posts_questions` q
INNER JOIN `bigquery-public-data.stackoverflow.posts_answers` a
ON q.id = a.parent_id
LIMIT 10;
```

## LEFT JOIN

This query returns all questions, even if they do not have answers.

```python
%%bigquery --project quiz11-joins

SELECT
q.id AS question_id,
q.title,
a.id AS answer_id
FROM `bigquery-public-data.stackoverflow.posts_questions` q
LEFT JOIN `bigquery-public-data.stackoverflow.posts_answers` a
ON q.id = a.parent_id
LIMIT 10;
```

## RIGHT JOIN

This query returns all answers and their corresponding questions.

```python
%%bigquery --project quiz11-joins

SELECT
q.id AS question_id,
q.title,
a.id AS answer_id
FROM `bigquery-public-data.stackoverflow.posts_questions` q
RIGHT JOIN `bigquery-public-data.stackoverflow.posts_answers` a
ON q.id = a.parent_id
LIMIT 10;
```

## FULL OUTER JOIN

This query returns all rows from both tables.

```python
%%bigquery --project quiz11-joins

SELECT
q.id AS question_id,
q.title,
a.id AS answer_id
FROM `bigquery-public-data.stackoverflow.posts_questions` q
FULL OUTER JOIN `bigquery-public-data.stackoverflow.posts_answers` a
ON q.id = a.parent_id
LIMIT 10;
```
